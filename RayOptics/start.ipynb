Now we need to import the package

```python
from raytracing import *
```

First we need to set up a path for the ray tracing to occur over.

```python
path = ImagingPath()
```

This path is just a 'thing' (a path) which we can hang things on
We are going to add some sapce, a lens and then some more space
To do this we are going to append things to the path, this is very similar to appending to a list in python.


```python
path.append(Space(d=150)) # This is the start of the simulation
path.append(Lens(f=50, diameter=20))  # This adds a lens with a focal length of 50 mm and a diameter of 20 mm
path.append(Space(d=150)) # After that I want 150 mm of space.
```

We can now look at our path with the display command

```python
path.display()
```

What does this show us.
Ff is the front focal length, Fb the back focal length.
The red, green and blue lines are fans, they are beams of light emerging from the objects top, middle and bottom.
The red arrow is the position of our image

While that look fun, it is a little annoying not being able to have a real object (or define one)
I am going to start a new path and again add a space, a lens and another space.
Now we are going to use the display with object command ( displayWithObject(size)

```python
path = ImagingPath()
path.append(Space(d=150)) # This is the start of the simulation
path.append(Lens(f=50, diameter=20))  # This adds a lens with a focal length of 50 mm and a diameter of 20 mm
path.append(Space(d=150)) # After that I want 150 mm of space.
path.displayWithObject(2)
```

We can chat about this in the lecture.
But for now lets put the object at the exact focal length of the lens

```python
path = ImagingPath()
path.append(Space(d=50)) # The object is now the same distance from the lens as the focal length
path.append(Lens(f=50, diameter=20))  # This adds a lens with a focal length of 50 mm and a diameter of 20 mm
path.append(Space(d=150)) # After that I want 150 mm of space.
path.displayWithObject(2)
```

We can see that each point would focus at infinity.
Next we will put the object 2 focal lengths from the lens
Note that we have set the focal length to 50

```python
focalLength = 50
objectHeight = 10
path = ImagingPath()
path.append(Space(d=focalLength*2)) # This is the start of the simulation
path.append(Lens(f=focalLength, diameter=20))  # This adds a lens with a focal length of 50 mm and a diameter of 20 mm
path.append(Space(d=150)) # After that I want 150 mm of space.
path.displayWithObject(objectHeight,fanAngle=30)
```
